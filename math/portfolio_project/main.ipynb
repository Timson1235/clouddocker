```python
import matplotlib.pyplot as plt
```

```python
# Sort the dataframe by 'Ladder score' in descending order
df_sorted = df.sort_values(by='Ladder score', ascending=False)

# Create a wider figure
fig, ax = plt.subplots(figsize=(20, 6))

# Plot the sorted data
ax.bar(df_sorted['Country name'], df_sorted['Ladder score'])
ax.set_xlabel('Country Name')
ax.set_ylabel('Happiness Score')
ax.set_title('Happiness Score by Country')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
```
