# Exercise 4. (Equation with Logarithms)

Solve for $x$, analytically and by plotting the solution in Python: $2 \log x = \log 2 + \log(3x - 4)$

For plotting the functions, use the numpy command `arange` to create a list of $x$-values.

$2 \log x = \log 2 + \log(3x - 4)$

$2 \log x = \log(6x - 8)$

0 = $\log(6x - 8) - 2 \log x$

0 = $\log(6x - 8) - \log x^2$

0 = $ \log ((6x-8)/x^2) $

Since $\log(1) = 0$, we set:
$\frac{6x - 8}{x^2} = 1$

$6x - 8 = x^2$

$x^2 - 6x + 8 = 0$

Using the quadratic formula:
$x = \frac{6 \pm \sqrt{36-32}}{2} = \frac{6 \pm \sqrt{4}}{2} = \frac{6 \pm 2}{2}$

So $x = 4$ or $x = 2$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 2 * np.log10(x) - np.log10(2) - np.log10(3*x - 4)

# Create x values using numpy arange
x_values = np.arange(2, 10, 0.1)  # Avoid x <= 0 and x < 4/3 (to prevent log errors)

# Compute y values
y_values = f(x_values)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label=r'$2 \log x - \log 2 - \log(3x - 4)$')
plt.axhline(0, color='black', linestyle='--', linewidth=1)  # x-axis
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Graphical Solution of the Logarithmic Equation")
plt.legend()
plt.grid()
plt.show()
```
