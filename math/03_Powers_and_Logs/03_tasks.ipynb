# Exercise 4. (Equation with Logarithms)

Solve for $x$, analytically and by plotting the solution in Python: $2 \log x = \log 2 + \log(3x - 4)$

For plotting the functions, use the numpy command `arange` to create a list of $x$-values.

$2 \log x = \log 2 + \log(3x - 4)$

$2 \log x = \log(6x - 8)$

0 = $\log(6x - 8) - 2 \log x$

0 = $\log(6x - 8) - \log x^2$

0 = $ \log ((6x-8)/x^2) $

Since $\log(1) = 0$, we set:
$\frac{6x - 8}{x^2} = 1$

$6x - 8 = x^2$

$x^2 - 6x + 8 = 0$

Using the quadratic formula:
$x = \frac{6 \pm \sqrt{36-32}}{2} = \frac{6 \pm \sqrt{4}}{2} = \frac{6 \pm 2}{2}$

So $x = 4$ or $x = 2$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 2 * np.log10(x) - np.log10(2) - np.log10(3*x - 4)

# Create x values using numpy arange
x_values = np.arange(2, 10, 0.1)  # Avoid x <= 0 and x < 4/3 (to prevent log errors)

# Compute y values
y_values = f(x_values)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label=r'$2 \log x - \log 2 - \log(3x - 4)$')
plt.axhline(0, color='black', linestyle='--', linewidth=1)  # x-axis
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Graphical Solution of the Logarithmic Equation")
plt.legend()
plt.grid()
plt.show()
```

# Exercise 5: Epidemic Spreading Model

Consider the following simplified model of epidemic spreading:

- For a disease, the value \( R \) determines the number of persons a sick person will infect on average.
- In every new step, each currently infected person infects \( R \) further persons and then stops being contagious.
- Assume that the disease is discovered when 100 patients are in their contagious state.

### Tasks:

1. **Mathematical Model:**  
   - Develop a function that models the number of new infections at each step.

2. **Plot the Spread:**  
   - Generate plots showing the number of new infections at each step for different values of \( R \).

3. **Qualitative Analysis:**  
   - Describe three distinct behaviors of the epidemic based on different values of \( R \).

4. **Sensitivity Analysis:**  
   - Examine the effects of small modifications to \( R \) (e.g., 1.7 vs. 1.75) on the spreading dynamics.


f(x) = 100*R^x

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, R):
    return 100 * R ** x  # Epidemic spread model

# Create x values
x_values = np.arange(0, 10, 0.1)  # Steps in the spreading process

def plot_graph(R):
    # Compute y values
    y_values = f(x_values, R)

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot(x_values, y_values, label=rf'$I_n = 100 \cdot {R}^x$', color='b')
    plt.xlabel("Steps (n)")
    plt.ylabel("New Infections (I_n)")
    plt.title(f"Epidemic Spread with R = {R}")
    plt.legend()
    plt.grid()
    plt.show()

# Example: Plot for R = 5
plot_graph(5)
```
