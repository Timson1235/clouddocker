# Gradient Descent for $y=x^2$

<div class="alert alert-block alert-info">
<b>Content:</b> In this short notebook, we try gradient descent for $y=x^2$. We hardcode the derivative and the gradient descent step.
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-5,5,0.1)
y=x**2
plt.plot(x,y)
```

```python
# do one step into the direction of the negative gradient using stepsize alpha
# x = current coord
# -2*x = derivative at current coord
# alpha = learning rate (step size)
def gradient_descent_step(x, alpha):
    return x-2*x*alpha
```

```python
x=3 # the x from where we start gradient descent
steps=[3]
for i in range(20):
#    x=gradient_descent_step(x, 1/6)
#    x=gradient_descent_step(x, 5/6)
#    x=gradient_descent_step(x, 1)
    x=gradient_descent_step(x, 2)
    steps.append(x)
#    print (x)
```

```python
x=np.arange(-5,5,0.1)
y=x**2
steps=np.array(steps)
plt.plot(x,y)
plt.plot(steps, steps*steps)
```
