```python
import pandas as pd
```

```python
tracks = pd.read_csv('data/tracks.csv')
lyrics = pd.read_csv('data/lyrics.csv')
artists = pd.read_json('data/artists.json')
charts_es = pd.read_excel('data/charts_es.xlsx')
charts_de = pd.read_excel('data/charts_de.xlsx')
charts_gb = pd.read_excel('data/charts_gb.xlsx')
charts_us = pd.read_excel('data/charts_us.xlsx')
```

```python
tracks.shape
```

```python
lyrics.head()
```

```python
artists.rename(columns={'genres':'artists_genres'}, inplace= True)
artists.rename(columns={'name':'artist'}, inplace= True)
artists.rename(columns={'id':'artist_ids'}, inplace= True)

artists.head()
```

```python
charts_de.head()
```

```python

# Melt the DataFrame to long format
charts_us_long = pd.melt(charts_us, id_vars=['variable'], var_name='song_index', value_name='value')
# Pivot to get the final long format DataFrame
charts_us_long_final = charts_us_long.pivot(index='song_index', columns='variable', values='value').reset_index()
charts_us_long_final['country'] = 'us'

charts_us_long_final.head()
```

```python

# Melt the DataFrame to long format
charts_gb_long = pd.melt(charts_gb, id_vars=['variable'], var_name='song_index', value_name='value')
# Pivot to get the final long format DataFrame
charts_gb_long_final = charts_gb_long.pivot(index='song_index', columns='variable', values='value').reset_index()
charts_gb_long_final['country'] = 'gb'


charts_gb_long_final.head()
```

```python

# Melt the DataFrame to long format
charts_es_long = pd.melt(charts_es, id_vars=['variable'], var_name='song_index', value_name='value')
# Pivot to get the final long format DataFrame
charts_es_long_final = charts_es_long.pivot(index='song_index', columns='variable', values='value').reset_index()
charts_es_long_final['country'] = 'es'


charts_es_long_final.head()
```

```python

# Melt the DataFrame to long format
charts_de_long = pd.melt(charts_de, id_vars=['variable'], var_name='song_index', value_name='value')
# Pivot to get the final long format DataFrame
charts_de_long_final = charts_de_long.pivot(index='song_index', columns='variable', values='value').reset_index()
charts_de_long_final['country'] = 'de'


charts_de_long_final.head()
```

Let the merging begin:

```python
all_charts = pd.concat([charts_de_long_final, charts_gb_long_final, charts_us_long_final, charts_es_long_final], ignore_index= True)
all_charts.rename(columns={'track_id':'id'}, inplace= True)

all_charts.shape
```

```python
print(lyrics.shape, tracks.shape, artists.shape, charts_de.shape)
```

```python
tracks.head()
```

```python
df_tl = pd.merge(left=tracks, right=lyrics, on=["id"], how='outer')
df_tla = pd.merge(left=df_tl, right=artists, on=["artist_ids"], how='outer')
df_tlac = pd.merge(left=df_tla, right=all_charts, on=["id"], how='outer')

df_tlac.sort_values(['artist_names','name'],inplace=True)

df_tlac.head()
```
