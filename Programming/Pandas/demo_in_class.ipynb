# Combining Data
- Merging
- Concatenation

```python
data1 = pd.read_csv("data/data1.csv")
print(data1.shape)
data1.head()
```

```python
data2 = pd.read_csv("data/data2.csv")
data2.head()
print(data2.shape)
```

```python
df = pd.merge(left=data1, right=data2, on=["iso3", "date"], how='inner')
print(df.shape)
df.head()
```

```python
df = pd.merge(left=data1, right=data2, on=["iso3", "date"], how='outer', indicator=True)
df.head()
```

```python
df._merge.value_counts()
```
