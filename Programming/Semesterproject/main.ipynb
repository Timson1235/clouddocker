### Second Games dataset:
The second dataset contains information about each instance of the olympic games from the year 1912 to the year 2032 with according information.

### Third Results dataset:
The third dataset is in a excel file with multiple sheets in the file each for a different year of the olympic games from the year 2002 until 2012.

```python
results_sheets = ['Year 2002', 'Year 2004', 'Year 2006', 'Year 2008', 'Year 2010', 'Year 2012']
results = {}

for sheet in results_sheets:
    df = pd.read_excel('results.xlsx', sheet_name=sheet, skiprows=1)
    df['year'] = sheet
    print(df.shape)
    results[sheet] = df


```

```python
results['Year 2002'].head()
```

### 2. Combine all provided data suitably into a single, tidy data set (e.g. by merging, concatenating or reshaping the data) such that you are able to answer the subsequent questions.

```python
all_results = pd.concat(results.values(), ignore_index=True)
all_results['year'] = all_results.year.transform(lambda x: int(x.split(' ')[1]))
all_results.rename(columns={'country': 'home_country'}, inplace=True)
all_results.drop(columns='country_code', inplace=True)
all_results.head()
```

```python
df = pd.merge(all_results, games, on = ['year', 'season'], how = 'left')

df.head()
```

```python
df_all = pd.merge(df, athletes, on = ['athlete_id'], how = 'outer', indicator=True)
```

```python
df_all['_merge'].value_counts()
```

```python
df = pd.merge(df, athletes, on = ['athlete_id'], how = 'left')
df.head()
```
