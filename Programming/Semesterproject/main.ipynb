# Exercise 3
1. Which are the 5 sports with the highest average age of gold medal winners? Provide the sport and
the average age. (Note: We define age as the number of completed years between the athlete’s
birth and the opening date of the olympic games.)


```python
df_gold = df[df['medal'] == 'Gold']

# Convert 'born' column to datetime
df_gold['born'] = pd.to_datetime(df_gold['born'])

# Convert 'opened' column to datetime
df_gold['opened'] = pd.to_datetime(df_gold['opened'])

# Calculate age as the difference between the opening date and the birth date
df_gold['age'] = (df_gold['opened'] - df_gold['born']).dt.days // 365

# Group by sport and calculate the average age
average_age_by_sport = df_gold.groupby('sport')['age'].mean().sort_values(ascending=False)

# Get the top 5 sports with the highest average age
top_5_sports = average_age_by_sport.head(5)

# Display the result
top_5_sports
```

```python
df_germany = df[(df.home_country == 'Germany') & (df.season == 'summer')]
df_germany = df_germany[['year', 'sport', 'event', 'medal']]
df_germany = df_germany[df_germany['medal'].notna()]
df_germany = df_germany.drop_duplicates()	
# df_germany.groupby('sport', 'year').count()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for the relevant years and group by year, sport, and medal type
df_germany = df_germany[df_germany['year'].isin([2004, 2008, 2012])]

# Create a pivot table: rows are 'sport', columns are 'medal', values are counts
df_pivot = df_germany.groupby(['year', 'sport', 'medal']).size().unstack(fill_value=0)

# Set up the subplots
fig, axes = plt.subplots(len(df_pivot.index.get_level_values('year').unique()), 1, figsize=(12, 12), sharex=True)

# Iterate over each year and plot stackable bars for that year
for i, year in enumerate(sorted(df_pivot.index.get_level_values('year').unique())):
    # Get the data for the current year
    df_year = df_pivot.loc[year]
    
    # Plot the stackable bars
    df_year[['Gold', 'Silver', 'Bronze']].plot(kind='bar', stacked=True, ax=axes[i], color=['gold', 'silver', 'brown'])

    axes[i].set_title(f'Germany Medal Count in {year}')
    axes[i].set_ylabel('Number of Medals')
    axes[i].tick_params(axis='x', rotation=90)  # Rotate x labels for better readability

axes[-1].set_xlabel('Sport')  # Set x-axis label on the last subplot
plt.tight_layout()
plt.show()
```
