For the following exercises manipulate img using array indexing and slicing and then
display the resulting image using plt.imshow():
 First 500 rows of the image

```python
# Slice the first 500 rows of the image
img_first_500_rows = img[:500]  # This slices the first 500 rows across all columns and color channels

# Display the manipulated image
plt.imshow(img_first_500_rows)
plt.title("First 500 Rows of the Image")
plt.axis('off') 
plt.show()
```

```python
# Slice the first 500 rows of the image
img_first_500_rows = img[:,-500:]  # This slices the first 500 rows across all columns and color channels

# Display the manipulated image
plt.imshow(img_first_500_rows)
plt.title("Last 500 columns of the Image")
plt.axis('off') 
plt.show()
```

```python
# Slice the first 500 rows of the image
img_first_500_rows = img[::2,:]  # This slices the first 500 rows across all columns and color channels

# Display the manipulated image
plt.imshow(img_first_500_rows)
plt.title("Every second row")
plt.axis('off') 
plt.show()
```

```python
# Zoom in appropriately to display just the head of the cutest alpaca
img_first_500_rows = img[:400,1000:1400]  # This slices the first 500 rows across all columns and color channels

# Display the manipulated image
plt.imshow(img_first_500_rows)
plt.title("cutest face")
plt.axis('off') 
plt.show()
```

```python
# Flip the image upside down and left to right
flipped_img = img[::-1,::-1]

# Display the flipped image
plt.imshow(flipped_img)
plt.title("Flipped Image (Upside Down and Left to Right)")
plt.axis('off')  # Hides the axes for a cleaner look
plt.show()
```

```python
#pixelate the image
# Set the pixel size (e.g., 10)
pixel_size = 10  

# Pixelate the image by slicing
pixelated_img = img[::pixel_size, ::pixel_size]  # This reduces the resolution by taking every 'pixel_size'-th pixel

# Display the pixelated image
plt.imshow(pixelated_img)
plt.title("Pixelated Image")
plt.axis('off')
plt.show()
```

```python
# Set all values that are smaller than 100 to 0 and display the resulting image.

# Create a writable copy of the image
img_copy = img.copy()

# Set all RGB values smaller than 100 to 0
img_copy[img_copy < 100] = 0

# Display the manipulated image
plt.imshow(img_copy)
plt.title("Image with RGB Values < 100 Set to 0")
plt.axis('off') 
plt.show()
```

```python
# Set all values of the green and blue channels to 0 and display the resulting image
img_copy[:,:,1:2] = 0

# Display the manipulated image
plt.imshow(img_copy)
plt.title("Image with RGB Values < 100 Set to 0")
plt.axis('off') 
plt.show()
```

Display both grades_df and grades. How do they differ?
Using your numpy array grades calculate the following statistics::
 What is the average grade of all students and tasks?
 What is the average grade per student?
 What is the average grade per task?
 What is the median grade per task?

```python
import pandas as pd

grades_df = pd.read_csv('grades.csv', index_col=0)
grades = grades_df.to_numpy()
```

```python
print(grades)
print(grades_df)

# Calculate statistics
average_grade = np.mean(grades)                  # Average grade of all students and tasks
average_grade_per_student = np.mean(grades, axis=1)  # Average grade per student
average_grade_per_task = np.mean(grades, axis=0)     # Average grade per task
median_grade_per_task = np.median(grades, axis=0)    # Median grade per task

# Display the statistics
print("\nStatistics:")
print(f"Average grade of all students and tasks: {average_grade:.2f}")
print(f"Average grade per student: {average_grade_per_student}")
print(f"Average grade per task: {average_grade_per_task}")
print(f"Median grade per task: {median_grade_per_task}")
```
