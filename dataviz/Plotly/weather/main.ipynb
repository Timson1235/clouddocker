```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("weather.csv")
df.head()
```

```python
df.columns
```

```python
df.groupby('name').sum()
```

```python
df_kiel = df[df['name'] == 'Kiel-Holtenau']
```

```python
# Plot temperature trends
fig_temp = px.line(
    df, 
    x='date', 
    y='temperature_air_mean_2m', 
    color='name', 
    title='Mean Air Temperature (2m) Over Time',
    labels={'temperature_air_mean_2m': 'Temperature (°C)', 'date': 'Date'},
    markers=True
)
fig_temp.update_layout(legend_title_text='Station ID')
fig_temp.show()
```

```python
# Filter data for January 2024
df_jan_2024 = df[(df["date"] > "2023-01-01") & (df["date"] < "2024-01-01")]

# Plot temperature trends for January 2024
fig_temp = px.line(
    df_jan_2024, 
    x='date', 
    y='temperature_air_mean_2m', 
    color='name', 
    title='Mean Air Temperature (2m) Over Time (January 2024)',
    labels={'temperature_air_mean_2m': 'Temperature (°C)', 'date': 'Date'},
    markers=True
)
fig_temp.update_layout(legend_title_text='Station ID')
fig_temp.show()
```

```python
fig_temp = px.line(
    df, 
    x='date', 
    y='snow_depth', 
    color='name', 
    title='snow_depth Over Time',
    labels={'snow_depth': 'snow_depth', 'date': 'Date'},
    log_y= True,
    markers=True
    
)
fig_temp.update_traces(marker={'size': 4})
fig_temp.update_layout(legend_title_text='Weatherstation')
fig_temp.show()
```
